```python
from mfnn_factory import create_np_data_sets, create_mfnn

#--- create data sets
training_data, validation_data, test_data = create_np_data_sets()
inputs, targets = training_data

#--- create neural network
mfnn = create_mfnn(input_size=inputs.shape[2])
mfnn.summary()

#--- train the neural network
loss_history, validation_loss_history = mfnn.fit(inputs, targets, 
                              learning_rate=0.1, epochs=600, batch_size=128, 
                              validation_data=validation_data)
```

```python
#--- display the losses on a graph

import matplotlib.pyplot as plt

plt.plot(loss_history, label='Training loss')
plt.plot(validation_loss_history, label='Validation loss')
plt.legend()
plt.ylim(ymax=0.5)        

```
